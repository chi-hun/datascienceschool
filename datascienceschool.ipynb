```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1,5)
lss = ['-','--','-.',':']
for df, ls in zip(dfs, lss):
  rv = sp.stats.chi2(df=df)
  plt.plot(xx, rv.pdf(xx), ls=ls, label = f'free {df}')
plt.legend()
plt.ylim(0, 0.6)
plt.show()
```

```python
np.random.seed(0)
x = sp.stats.beta(15,12).rvs(10000)
sns.distplot(x, kde=False, norm_hist=True)
plt.show()
```

```python
def find_beta(x):
  x_bar = x.mean()
  s2 = x.var()
  a = x_bar * (x_bar * (1 - x_bar) / s2 - 1)
  b = (1 - x_bar) * (x_bar * (1 - x_bar) / s2 - 1)
  return a, b
t = find_beta(x)
print(t)
```

```python
xx = np.linspace(0,1,1000)
sns.distplot(x, kde=False, norm_hist=True)
plt.plot(xx, sp.stats.beta(t[0],t[1]).pdf(xx))
plt.show()
```

```python
mu = np.linspace(-3, 3, 1000)
sigma2 = 1
def likelihood(mu):
    return (2 * np.pi * sigma2) ** (3 / 2) * np.exp(-(3 * mu ** 2 + 4 * mu + 10) / (2 * sigma2))
li = likelihood(mu)
plt.plot(mu, li)
plt.show()
```

```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
mu = 0.65
a, b = 1, 1
xx = np.linspace(0,1,1000)
plt.plot(xx, sp.stats.beta(a, b).pdf(xx), ls=':')
plt.subplot(2,1,1)
for i in range(3):
  x = sp.stats.bernoulli(mu).rvs(50)
  n0, n1 = np.bincount(x, minlength=2)
  a, b= a+n1, b+n0
  plt.plot(xx, sp.stats.beta(a,b).pdf(xx), label=f'{i}')
plt.vlines(x=0.65, ymin=0, ymax=12)
plt.ylim(0, 12)
plt.legend()
plt.subplot(2,1,2)
for i in range(3):
  x = sp.stats.bernoulli(mu).rvs(1000)
  n0, n1 = np.bincount(x, minlength=2)
  a, b= a+n1, b+n0
  plt.plot(xx, sp.stats.beta(a,b).pdf(xx), label=f'{i}')
plt.vlines(x=0.65, ymin=0, ymax=50)
plt.ylim(0, 50)
plt.legend()
plt.show()
```

```python
N_1 = 50
N_1_1 = 100
mu_1 = 0
sigma_1 = 1
N_2 = 50
N_2_1 = 100
mu_2 = 0.5
sigma_2 = 1
plt.subplot(1,2,1)
plt1 = sp.stats.norm(mu_1, sigma_1).rvs(N_1)
plt2 = sp.stats.norm(mu_2, sigma_2).rvs(N_2)
sns.distplot(plt1, kde=False, fit=sp.stats.norm, label="num 1")
sns.distplot(plt2, kde=False, fit=sp.stats.norm, label="num 2")
plt.subplot(1,2,2)
plt3 = sp.stats.norm(mu_1, sigma_1).rvs(N_1_1)
plt4 = sp.stats.norm(mu_2, sigma_2).rvs(N_2_1)
sns.distplot(plt3, kde=False, fit=sp.stats.norm, label="num 1")
sns.distplot(plt4, kde=False, fit=sp.stats.norm, label="num 2")
plt.legend()
plt.show()
print(np.mean(plt1), np.mean(plt2))
print(sp.stats.ttest_ind(plt1, plt2, equal_var=False))
print(np.mean(plt3), np.mean(plt4))
print(sp.stats.ttest_ind(plt3, plt4, equal_var=False))
```
