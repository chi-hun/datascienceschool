```python
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
mu = 0
std = 1
rv = sp.stats.norm(mu, std)
xx = np.linspace(-5, 5, 100)
plt.subplot(2,1,1)
plt.plot(xx, rv.pdf(xx))
plt.subplot(2,1,2)
x = rv.rvs(20)
sns.distplot(x, rug=True, kde=False, fit=sp.stats.norm)
plt.show
print(np.mean(x), np.std(x))
```

```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1,5)
lss = ['-','--','-.',':']
for df, ls in zip(dfs, lss):
  rv = sp.stats.chi2(df=df)
  plt.plot(xx, rv.pdf(xx), ls=ls, label = f'free {df}')
plt.legend()
plt.ylim(0, 0.6)
plt.show()
```

```python
np.random.seed(0)
x = sp.stats.beta(15,12).rvs(10000)
sns.distplot(x, kde=False, norm_hist=True)
plt.show()
```

```python
def find_beta(x):
  x_bar = x.mean()
  s2 = x.var()
  a = x_bar * (x_bar * (1 - x_bar) / s2 - 1)
  b = (1 - x_bar) * (x_bar * (1 - x_bar) / s2 - 1)
  return a, b
t = find_beta(x)
print(t)
```

```python
xx = np.linspace(0,1,1000)
sns.distplot(x, kde=False, norm_hist=True)
plt.plot(xx, sp.stats.beta(t[0],t[1]).pdf(xx))
plt.show()
```

```python
mu = np.linspace(-3, 3, 1000)
sigma2 = 1
def likelihood(mu):
    return (2 * np.pi * sigma2) ** (3 / 2) * np.exp(-(3 * mu ** 2 + 4 * mu + 10) / (2 * sigma2))
li = likelihood(mu)
plt.plot(mu, li)
plt.show()
```

```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
mu = 0.65
a, b = 1, 1
xx = np.linspace(0,1,1000)
plt.plot(xx, sp.stats.beta(a, b).pdf(xx), ls=':')
for i in range(3):
  x = sp.stats.bernoulli(mu).rvs(50)
  n0, n1 = np.bincount(x, minlength=2)
  a, b= a+n0, b+n1
  plt.plot(xx, sp.stats.beta(a,b).pdf(xx), label=f'{i}')
plt.vlines(x=0.65, ymin=0, ymax=12)
plt.ylim(0, 12)
plt.legend()
plt.show()
```
