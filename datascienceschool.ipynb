```python
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
mu = 0
std = 1
rv = sp.stats.norm(mu, std)
xx = np.linspace(-5, 5, 100)
plt.subplot(2,1,1)
plt.plot(xx, rv.pdf(xx))
plt.subplot(2,1,2)
x = rv.rvs(20)
sns.distplot(x, rug=True, kde=False, fit=sp.stats.norm)
plt.show
print(np.mean(x), np.std(x))
```

```python
import pandas_datareader.data as web

symbols = ['SP500', 'NASDAQCOM', 'DJCA', 'NIKKEI225']
data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred')[sym]
data = data.dropna()
(data / data.iloc[0] * 100).plot()
plt.rc('font', family='Malgun Gothic')
plt.ylabel("")
plt.ylabel(" ")
plt.show()
```

```python
xx = np.linspace(-4,4,100)
for df in [1,2,5,10,20]:
  rv = sp.stats.t(df=df)
  plt.plot(xx, rv.pdf(xx), label=("dof=%d" % df))
plt.plot(xx, sp.stats.norm.pdf(xx), label='norm')
plt.legend()
plt.show
```
